```python
import pandas as pd
import numpy as np
from datetime import datetime
```

## Load Data

```python
file_path = "../dataset/data_inovasi.xlsx"
df = pd.read_excel(file_path)
df.head()
```

```python
print("Nilai unik di 'Koordinat':", df['Koordinat'].unique())
```

## Checking

```python
df.info()
```

## Missing Values

```python
df['Urusan lain yang beririsan'] = df['Urusan lain yang beririsan'].fillna('Tidak ada')
df['Urusan lain yang beririsan'] = df['Urusan lain yang beririsan'].fillna('-')
df['Link Video'] = df['Link Video'].fillna('-')
```

## Duplicates

```python
duplicates = df[df.duplicated(subset=['Judul Inovasi', 'Admin OPD'], keep=False)]
if not duplicates.empty:
    print("Duplikat ditemukan:")
    print(duplicates[['No', 'Judul Inovasi', 'Admin OPD']])
    df = df.drop_duplicates(subset=['Judul Inovasi', 'Admin OPD'], keep='first')
    print(f"\nDuplikat dihapus. Jumlah baris sekarang: {len(df)}")
```

## Text Cleaning

```python
string_columns = ['Judul Inovasi', 'Pemda', 'Admin OPD', 'Inisiator', 'Nama Inisiator', 
                  'Bentuk Inovasi', 'Jenis', 'Asta Cipta', 'Urusan Utama', 'Urusan lain yang beririsan', 
                  'Tahapan Inovasi', 'Video', 'Link Video']
for col in string_columns:
    df[col] = df[col].astype(str).str.strip()

df['Bentuk Inovasi'] = df['Bentuk Inovasi'].str.title()
df['Urusan Utama'] = df['Urusan Utama'].str.title()
df['Jenis'] = df['Jenis'].str.title()
df['Tahapan Inovasi'] = df['Tahapan Inovasi'].str.title()
df['Video'] = df['Video'].str.title()
df.head()
```

## Datetime

```python
date_columns = ['Tanggal Input', 'Tanggal Penerapan', 'Tanggal Pengembangan']
for col in date_columns:
    df[col] = df[col].replace('-', pd.NaT)
    df[col] = pd.to_datetime(df[col], format='%d-%m-%Y', errors='coerce')
    df[col] = df[col].dt.strftime('%Y-%m-%d').where(df[col].notna(), None)  
```

## Coordinates

```python
def split_coordinates(coord):
    if pd.isna(coord) or coord == '-':
        return pd.Series([np.nan, np.nan])
    try:
        lat, lon = coord.split(', ')
        return pd.Series([float(lat), float(lon)])
    except:
        return pd.Series([np.nan, np.nan])

df[['Latitude', 'Longitude']] = df['Koordinat'].apply(split_coordinates)
df = df.drop(columns=['Koordinat'])
```

## Check Unique Value

```python
print("Nilai unik di 'Inisiator':", df['Inisiator'].unique())
print("Nilai unik di 'Bentuk Inovasi':", df['Bentuk Inovasi'].unique())
print("Nilai unik di 'Jenis':", df['Jenis'].unique())
print("Nilai unik di 'Tahapan Inovasi':", df['Tahapan Inovasi'].unique())
print("Nilai unik di 'Video':", df['Video'].unique())
print("Nilai unik di 'Kematangan':", df['Kematangan'].unique())
print("Nilai unik di 'Urusan Utama':", df['Urusan Utama'].unique())
print("Nilai unik di 'Longitude':", df['Asta Cipta'].unique())
print("Nilai unik di 'Longitude':", df['Urusan lain yang beririsan'].unique())
```

## Change Column Name

```python
df.columns = df.columns.str.lower().str.replace(' ', '_').str.replace('.', '')
```
