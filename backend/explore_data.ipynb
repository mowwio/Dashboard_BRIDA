```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data_inovasi_clean.csv')
df
```

```python
print("Shape data:", df.shape)
```

```python
print(df.info())
```

```python
print(df.describe(include='all'))
```

```python
categorical_cols = ['inisiator', 'bentuk_inovasi', 'jenis', 'tahapan_inovasi', 'video']

for col in categorical_cols:
    print(f"\nDistribusi {col}:")
    print(df[col].value_counts())
    plt.figure(figsize=(10, 5)) 
    df[col].value_counts().plot(kind='barh', color='skyblue', edgecolor='black')
    plt.title(f'Distribusi {col.replace("_", " ").title()}')
    plt.xlabel('Jumlah Data') 
    plt.gca().invert_yaxis() 
    plt.show()
```

```python
numeric_cols = ['kematangan', 'latitude', 'longitude']
for col in numeric_cols:
    if col in df.columns:
        print(f"\nStatistik {col}:")
        print(df[col].describe())
        
        # Histogram
        plt.figure(figsize=(8, 4))
        sns.histplot(df[col].dropna(), bins=20, kde=True, color='green')
        plt.title(f'Distribusi {col.title()}')
        plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='tahapan_inovasi', y='kematangan', palette='Set2')
plt.title('Kematangan berdasarkan Tahapan Inovasi')
plt.xticks(rotation=45)
plt.show()

# Scatter plot latitude vs longitude (untuk lokasi)
if 'latitude' in df.columns and 'longitude' in df.columns:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df.dropna(subset=['latitude', 'longitude']), x='longitude', y='latitude', hue='bentuk_inovasi', palette='viridis')
    plt.title('Lokasi Inovasi berdasarkan Bentuk')
    plt.show()
```

```python
date_cols = ['tanggal_input', 'tanggal_penerapan', 'tanggal_pengembangan']
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')

# Jumlah inovasi per bulan (misalnya, berdasarkan tanggal_input)
df['bulan_input'] = df['tanggal_input'].dt.to_period('M')
monthly_count = df.groupby('bulan_input').size()
plt.figure(figsize=(10, 5))
monthly_count.plot(kind='line', marker='o', color='orange')
plt.title('Jumlah Inovasi per Bulan (Tanggal Input)')
plt.xticks(rotation=45)
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
if not numeric_df.empty:
    plt.figure(figsize=(8, 6))
    sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
    plt.title('Korelasi Antar Kolom Numerik')
    plt.show()
```

```python
top_inovasi = df.nlargest(10, 'kematangan')[['judul_inovasi', 'kematangan', 'inisiator']]
print("Top 10 Inovasi berdasarkan Kematangan:")
print(top_inovasi)
```

```python
avg_kematangan = df.groupby('inisiator')['kematangan'].mean().sort_values(ascending=False)
print("Rata-rata Kematangan per Inisiator:")
print(avg_kematangan)
```
